## So sánh kết quả với thư viện

```python
regr = LinearRegression()
regr.fit(train_x, train_y)
print("Coefficients : ", regr.coef_)
print("Intercepts : ", regr.intercept_)
```

```python
def check_with_sklearn(test_x, test_y, name_GD, w, b):

  print(f"This is the result of {name_GD} : ")
  test_y_pre = regr.predict(test_x)

  print("\tMean absolute error: %.2f" % np.mean(np.absolute(test_y_pre - test_y)))
  print("\tResidual sum of squares (MSE): %.2f" % np.mean((test_y_pre - test_y) ** 2))
  print("\tR2-score: %.2f" % r2_score(test_y , test_y_pre))
  test_y_ = predict(test_x, w, b)

  print("\tMean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
  print("\tResidual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
  print("\tR2-score: %.2f" % r2_score(test_y , test_y_))
  print("\tCost: %.2f" % compute_cost(test_y , test_y_) )
```
